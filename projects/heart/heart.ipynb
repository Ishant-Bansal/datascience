```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('heart.csv')
df = pd.DataFrame(data)
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()   
```

```python
df.dtypes
```

```python
df['Sex'].unique()
```

```python
df['Sex'] = df['Sex'].map({'M': 1, 'F': 0})
```

```python
df.dtypes
```

```python
df['ChestPainType'].unique()
```

```python
df['ChestPainType'] = df['ChestPainType'].map({'ATA' : 1,
                                               'NAP' : 2, 
                                               'ASY' : 3, 
                                               'TA' : 4})
```

```python
df.dtypes
```

```python
df['RestingECG'].unique()
```

```python
df['RestingECG'] = df['RestingECG'].map({'Normal' : 1, 
                                       'ST' : 2, 
                                       'LVH' : 3})
```

```python
df.dtypes
```

```python
df['ExerciseAngina'].unique()
```

```python
df['ExerciseAngina'] = df['ExerciseAngina'].map({'Y' : 1,
                                                'N' : 0})
```

```python
df.dtypes
```

```python
df['ST_Slope'].unique()
```

```python
df['ST_Slope'] = df['ST_Slope'].map({'Up' : 1,
                                    'Flat' : 2,
                                    'Down' : 3})
```

```python
df.dtypes
```

```python
x = df.drop(['HeartDisease'], axis=1)
y = df['HeartDisease']
```

```python
x
```

```python
y
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
rfr = RandomForestRegressor()
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
rfr.fit(x_train, y_train)
```

```python
rfr.score(x_test, y_test)
```

```python
rfr.predict(x_test)
```
