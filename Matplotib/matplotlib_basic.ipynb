```python
x_point = np.array([1,2,3,4,5,])
y_point = np.array([2,4,6,2,10])
plt.title("lineplot" ,fontsize = '20',fontweight ='bold' , color = 'pink' )
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.plot(x_point,linestyle = 'dotted')
plt.plot(y_point,linestyle = 'dotted')
```

```python
x = np.array([1,2,3,4,5,])
y = np.array([2,4,6,2,10])
c=np.array(['red','green','blue','yellow','pink'])
s = np.array([50,60,70,80,90])
plt.scatter(x,y,c=c,s=s)
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.scatter(x, y,alpha=0.5, s = sizes, c = c,cmap = 'viridis') # camap = color map, viridis, nipy_spectral, coolwarm
plt.colorbar()
plt.show()
```
